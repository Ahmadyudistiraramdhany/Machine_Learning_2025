```python
import pandas as pd

# 1. Baca dataset
df = pd.read_csv('../data/day.csv')

# 2. Acak data (agar tidak bias urutan)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# 3. Hitung jumlah total data
total_data = len(df)

# 4. Hitung jumlah untuk masing-masing bagian
train_size = int(0.8 * total_data)
test_size = total_data - train_size
val_size = int(0.1 * train_size)  # 10% dari training

# 5. Bagi dataset
train_df = df[:train_size]
test_df = df[train_size:]

# Bagi validation dari train_df
val_df = train_df[:val_size]
train_df = train_df[val_size:]  # sisa untuk training

# 6. Tampilkan hasil
print(f"Jumlah data    : {(total_data)}")
print(f"Training set   : {len(train_df)}")
print(f"Validation set : {len(val_df)}")
print(f"Testing set    : {len(test_df)}")

print("\n5 Baris pertama dari Training set:")
print(train_df.head())

print("\n5 Baris pertama dari Validation set:")
print(val_df.head())

print("\n5 Baris pertama dari Testing set:")
print(test_df.head())
```
